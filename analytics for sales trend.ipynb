```python
import numpy as np
import pandas as pd
import math
```

```python
df=pd.read_csv("1-sales_performance_report.csv")
df.head(15)
```

```python
actual=df['Actual_Sales']
actual.head()
```

```python
mov_avg=[]
sum=0
count=0
for i in range(len(actual)):
    while count<=7:
        sum+=actual[i]
        count+=1
        break
    if count==7:
        ele=sum/count
        mov_avg.append(ele)
        sum=0
        count=0
print(mov_avg)

```

```python
df1=pd.DataFrame(mov_avg)
```

```python
df=pd.concat([df,df1],axis=1)
```

```python
df.head()
```

```python
df.columns=['Date','Product','Actual_Sales','Sales_Target','7Day_Moving_Average']
```
