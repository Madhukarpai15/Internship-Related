```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Sample sales data for U2.csv')
df.head
```

```python
actual=df['Actual_Sales']
target=df['Sales_Target']
```

```python
actual.head
```

```python
target.head
```

```python
import math
error=[]
for i in range(len(actual)):
    ele=actual[i]-target[i]
    error.append(ele)
print(error)
```

```python
abs_error=[]
for i in range(len(error)):
    ele=math.fabs(error[i])
    abs_error.append(ele)
print(abs_error)
```

```python
perc_error=[]
for i in range(len(abs_error)):
    ele=(abs_error[i]/actual[i])*100
    perc_error.append(ele)
print(perc_error)
```

```python
sq_error=[]
for i in range(len(error)):
    ele=error[i]**2
    sq_error.append(ele)
print(sq_error)
```
