```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Assignment 1 Dataset.csv')
df.head()
```

```python
actual=df['Actual_Sales']
target=df['Forecasted_Sales']
```

```python
actual.head()
```

```python
target.head()
```

```python
import math
error=[]
for i in range(len(actual)):
    ele=actual[i]-target[i]
    error.append(ele)
print(error)
```

```python
abs_error=[]
for i in range(len(error)):
    ele=math.fabs(error[i])
    abs_error.append(ele)
print(abs_error)
```

```python
perc_error=[]
for i in range(len(abs_error)):
    ele=(abs_error[i]/actual[i])*100
    perc_error.append(ele)
print(perc_error)
```

```python
sq_error=[]
for i in range(len(error)):
    ele=error[i]**2
    sq_error.append(ele)
print(sq_error)
```

```python
df1=pd.DataFrame(error)
df2=pd.DataFrame(abs_error)
df3=pd.DataFrame(perc_error)
df4=pd.DataFrame(sq_error)

```

```python
df=pd.concat([df,df1,df2,df3,df4],axis=1)
```

```python
df.columns=['Date','Product','Actual_Sales','Sales_Target','Error','Absolute Error','Percentage Error','Squared Error']
```

```python
df.head()
```

```python
mape_val=[]
for i in range(len(actual)):
    ele=math.fabs((actual[i]-target[i])/actual[i])*100
    mape_val.append(ele)
print(mape_val)
```

```python
Sum=0
for i in mape_val:
    Sum+=i
mape=Sum/len(mape_val)
print(mape)
```

```python
MAPE=[mape]
df5=pd.DataFrame(MAPE)
df=pd.concat([df,df5],axis=1)
```

```python
mad_val=[]
for i in range(len(actual)):
    ele=math.fabs(actual[i]-target[i])
    mad_val.append(ele)
print(mad_val)
```

```python
summ=0
for i in mad_val:
    summ+=i
mad=Sum/len(mad_val)
print(mad)
```

```python
MAD=[mad]
df6=pd.DataFrame(MAD)
df=pd.concat([df,df6],axis=1)
```

```python
fore_val=[]
for i in range(len(target)):
    ele=target[i]-actual[i]
    fore_val.append(ele)
print(fore_val)
```

```python
ssum=0
for i in fore_val:
    ssum+=i
fore=ssum/len(fore_val)
print(fore)
```

```python
BIAS=[fore]
df7=pd.DataFrame(BIAS)
df=pd.concat([df,df7],axis=1)
```

```python
sse=df["Squared Error"]
suum=0
for i in sse:
    suum+=i
var=suum/len(sse)
print(var)
```

```python
VAR=[var]
df8=pd.DataFrame(VAR)
df=pd.concat([df,df8],axis=1)
```
